```python
import pandas as pd

df = pd.read_parquet('data/00000000.parquet')

df = df.sort_values('duration', ascending=False)

df.head(n=10)
```

```python
for index, row in df.head(10).iterrows():
    with open(f'data/task_1/audio_{index}.wav', mode='bx') as f:
        f.write(row['audio']['bytes'])
    with open(f'data/task_1/transcript_{index}.txt', mode='w') as f:
        f.write(row['text'])
```

```python
import wave

def get_audio_info(filename):
    with wave.open(filename, 'rb') as audio_file:
        sample_rate = audio_file.getframerate()
        num_channels = audio_file.getnchannels()

        if num_channels == 1:
            channel_type = "Mono"
        elif num_channels == 2:
            channel_type = "Stereo"
        else:
            channel_type = "Other"

        return {
            "sample_rate": sample_rate,
            "channels": num_channels,
            "channel_type": channel_type
        }

file_path = "data/task_1/audio_8.wav"
info = get_audio_info(file_path)
print(f"Sample Rate: {info['sample_rate']} Hz")
print(f"Channels: {info['channels']} ({info['channel_type']})")
```
